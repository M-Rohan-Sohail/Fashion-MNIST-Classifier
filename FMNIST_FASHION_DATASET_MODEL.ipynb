```python
#Importing the Libs
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
fmnist=tf.keras.datasets.fashion_mnist
```

```python
(training_data,training_labels),(testing_data,testing_labels)=fmnist.load_data() #Loading the Dataset
```

```python
#Normalizing the Images
training_data=training_data/255
testing_data=testing_data/255
#These scacles images between 0,1
```

```python
#Building the Classification Model

model=tf.keras.models.Sequential([
    tf.keras.Input(shape=(28,28)), # As your images are in 28x28 pixels
    tf.keras.layers.Flatten(), #converts our images to 1D arrays    
    tf.keras.layers.Dense(128,activation=tf.nn.relu), # relu is means if input is >x thenn return x otherwise return 0
    tf.keras.layers.Dense(10,activation=tf.nn.softmax) # take the inputs and make there sum=1 thats mean it makes the proability 
])

# The 128 are the hidden neurons that are used to learn the complex patterns means creating 128 features for learning from the input
# The 10 means the number of the classes on which the model make predictions
```

```python
#Compiling the Models
model.compile(optimizer=tf.optimizers.Adam(),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]      
              )
model.fit(training_data,training_labels,epochs=100)
```

```python
#Evaluating the Results

model.evaluate(testing_data,testing_labels)
```
